# Cosine Wave FFT
Let's generate a cosine wave of **cycles** cycles, sampled over **time_range** seconds, then do a FFT on this to find the frequency responses.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Create a number of cosine waves

```python
samples = 1024
cycles = 10
x = np.linspace(0, 2*np.pi*cycles, samples)
y = np.cos(x)
```

```python
cycles = 50
x = np.linspace(0, 2*np.pi*cycles, samples)
y += 0.5*np.cos(x)
```

Range from 0 to 1s

```python
time_range = 1
x_sec = np.linspace(0, time_range, samples)
```

```python
plt.plot(x_sec, y)
plt.title('Cosine Wave - 1sec')
```

```python
yf = np.fft.fft(y)
```

The frequency is divided up into *samples* quanta - but don't forget to factor in how much time we sampled for. 

We only plot half of the samples.  We divide the FFT result by samples * 0.5 to get magnitude of the freq response.  

```python
x_freq = np.linspace(0, 1.0/time_range * samples, samples)
yf_mag = 2.0/samples * np.abs(yf)
plt.plot(x_freq[0:samples/2], yf_mag[0:samples/2])
plt.grid()
plt.xlabel("Frequency Hz")
plt.ylabel("Magnitude")
```

```python
yf_dec = 20.0 * np.log10(yf_mag)
plt.plot(x_freq[0:samples/2], yf_dec[0:samples/2])
plt.grid()
plt.xlabel("Frequency Hz")
plt.ylabel("Decibels")
```
